```python
plt.figure(figsize=(11, 5))
p1 = sns.heatmap(
    df,
    yticklabels=['Gradient Boosting', 'Random Forest', 'Decision Tree', 'KNN', 'SVM', 'Logistical'],
    linewidth=.5,
    fmt=".1f",
    cmap="crest",
    square=True
)
p1.set(xlabel="Fold", ylabel="Classifier", title="Variations on Accuracy per Classifier Across Multi-fold Validation")
plt.savefig('heatmap.png')
```

```python
plt.figure(figsize=(20, 20))
df['mean'] = df.mean(axis=1)
df['min'] = df.min(axis=1)
df['max'] = df.max(axis=1)
aggregate_df = pd.DataFrame({'Classifier': ['Gradient Boosting', 'Random Forest', 'Decision Tree', 'KNN', 'SVM', 'Logistical'], 'Minimum': df['min'], 'Mean': df['mean'], 'Maximum': df['max']})
ax = aggregate_df.plot.bar(x='Classifier', y='Mean', rot=0, ylim=(70, 82))
ax.set(xlabel="Classifier", ylabel="Average Accuracy", title="Mean Accuracy per Classifier")
plt.xticks(rotation=-45)
ax.get_legend().remove()
plt.savefig('average_accuracy.png')
```
